```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
embeddings= OpenAIEmbeddings(model="text-embedding-3-large", dimensions=300)
```

```python
docs=[
    "The Pacific Ocean is the largest and deepest of the world’s oceans, covering more than 63 million square miles. It extends from the Arctic Ocean in the north to the Southern Ocean in the south and is bounded by Asia and Australia on the west and the Americas on the east.",
    "The quarterly financial report showed a 12% increase in net revenue, driven largely by increased customer acquisition and higher subscription retention rates in the Asia-Pacific region. Operational costs decreased by 5%, boosting overall profitability.",
    "FastAPI is a modern Python web framework for building APIs with automatic OpenAPI schema generation. It supports asynchronous programming using async/await syntax and is built on top of Starlette and Pydantic for performance and validation.",
    "Hypertension, commonly known as high blood pressure, is a condition in which the force of the blood against artery walls is consistently too high. It can lead to serious health problems like heart disease, stroke, and kidney failure if left untreated.",
    "The experiment demonstrated that under low-temperature conditions, the reaction rate of the enzyme catalase increased significantly when exposed to a mildly acidic pH level of 6.2. These findings support previous research indicating enzyme activity is pH-sensitive.",
    "According to the GDPR regulation enacted in 2018, organizations handling EU citizens' data must obtain explicit consent, ensure data portability, and provide the right to be forgotten. Non-compliance can result in hefty fines and legal actions."
]
```

```python
docs_embedding= embeddings.embed_documents(docs)
```

```python
query="where cricket watch the most?"
```

```python
query_embedding= embeddings.embed_query(query)
```

```python
similarity_scores=cosine_similarity([query_embedding], docs_embedding)[0]
```

```python
similarity_scores
```

```python
index, score= sorted(list(enumerate(similarity_scores)), key= lambda x: x[1], reverse=True)[0]
```

```python
print("User Asked Query: ", query)
print("documents index", docs[index])
print("similarity score", score)
```
